```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

import scienceplots  
plt.style.use(['science','notebook', 'grid'])

from matplotlib import rc
rc('text', usetex=True)
font = {'family' : 'serif',
        'weight' : 'bold',
        'size'   : '14'}
rc('font', **font)
```

```python
# voltage 
V = np.array([0.0214, 3.0029, 4.9948, 6.9937, 10.0732, 12.0001, 14.9921])
# Current 
I = np.array([4.2149, 2.4747, 1.3047, 0.1479, -1.4814, -2.7531, -4.5016])
```

```python
x = V
y = I

coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef) 
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x,y, 'bo', x, poly1d_fn(x), '--k') #'--k'=black dashed line, 'yo' = yellow circle marker
plt.xlabel(r'Voltage $V_2$ [V]')
plt.ylabel(r'Current I [A]')
# plt.text(0.55, 0.85, '$y = (0.21 \pm 0.07)x + (1.8 \pm 1.9)$', transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='red'))
plt.savefig('current.vs.voltage.png')
plt.show()
```
