```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import pandas as pd

import scienceplots  
plt.style.use(['science','notebook', 'grid'])

from matplotlib import rc
rc('text', usetex=True)
font = {'family' : 'serif',
        'weight' : 'bold',
        'size'   : '14'}
rc('font', **font)
```

### Phase Plot

```python
df = pd.read_excel('Electronic_Filters_Plots.xlsx')

data_array = df.to_numpy()
transposed_array = data_array.transpose()
# x: Frequency
x = np.array(transposed_array[0])
# y: Phase
y = np.array(transposed_array[1])
```

```python
plt.plot(x, y, 'bo')
plt.xlabel("Frequency [Hz]")
plt.ylabel("Phase (degrees)")
plt.savefig("phase_plot.png")
plt.show()
```

### Bode Plot

```python
# assigning the variables from the .xlsx file
# x1: log(f/f_co) where f_co is the cutoff frequency
x1 = np.array(transposed_array[2])
y1 = np.array(transposed_array[3])
```

```python
plt.plot(x1, y1, 'go')
plt.xlabel(r"$\log \left( \frac{f}{f_{co}}\right)$")
plt.ylabel("Power Gain [dB]")
plt.savefig("bode_plot.png")
plt.show()
```
