```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

import scienceplots  
plt.style.use(['science','notebook', 'grid'])

from matplotlib import rc
rc('text', usetex=True)
font = {'family' : 'serif',
        'weight' : 'bold',
        'size'   : '14'}
rc('font', **font)
```

### Imaging the graph for the case with 2 wires

```python
# x-data for 2-wire apparatus
# length/Area
x = np.array([7894100.966,15788201.93,23682302.9,31576403.86,39470504.83,24867959.86])
# y-data for 2-wire apparatus
# Resistivity
y = np.array([2.06,4.65,7.07,9.50,11.88,6.72])
```

```python
coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef)
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x,y, 'bo', x, poly1d_fn(x), '--k') #'--k'=black dashed line, 'yo' = yellow circle marker
ax.set_xlabel(r'$\frac{L}{A}$ $[m^{-1}]$')
ax.set_ylabel(r'Resistance [$\Omega$]')
ax.text(0.05, 0.85, r'$y = (3.1 \pm 0.1) \times 10^{-7} x + (-0.4 \pm 0.3)$', transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='red'))
plt.savefig('2wire.png')
plt.show()
```
